```python
# Dependencies and Setup
import pandas as pd
import scipy.stats as st

# Study data files
men_path = "data/Men_workers.csv"

# Read the mouse data and the study results
men_results = pd.read_csv(men_path)
men_results
```

```python
men_workers = pd.DataFrame(men_results)
men_workers_df = men_workers.rename(columns={"Number\nof\nworkers": "Total Number of Males",
                                                 "Median\nweekly\nearnings": "Median Earnings(total)"}).drop(0)

#men_workers_df.dtypes

men_numbers = men_workers_df["Total Number of Males"].str.replace(",","").astype(int)

new_men_df = men_workers_df.drop(columns = ['Total Number of Males'])
new_men_df

men_numbers_df = pd.DataFrame(men_numbers)
men_numbers_df

final_men_df = pd.concat([new_men_df, men_numbers_df], 1)
final_men_df

#sorted_df = final_men_df.sort_values(by=["Total Number of Men"], ascending = False)

male_numbers_df = final_men_df.drop(['Median Earnings(total)', 'Standard error of median'], axis=1)
```

```python
# Dependencies and Setup
import pandas as pd
import scipy.stats as st

# Study data files
Female_path = "data/Female_workers.csv"

# Read the mouse data and the study results
Female_results = pd.read_csv(Female_path)
Female_results
```

```python
female_workers = pd.DataFrame(Female_results)
female_workers_df = female_workers.rename(columns={"Number\nof\nworkers": "Total Number of Females",
                                                 "Median\nweekly\nearnings": "Median Earnings(total)"}).drop(0)

#men_workers_df.dtypes

female_numbers = female_workers_df["Total Number of Females"].str.replace(",","").astype(int)

new_female_df = female_workers_df.drop(columns = ['Total Number of Females'])
new_female_df

female_numbers_df = pd.DataFrame(female_numbers)
female_numbers_df

final_female_df = pd.concat([new_female_df, female_numbers_df], 1)
final_female_df

female_numbers_df = final_female_df.drop(['Occupation','Median Earnings(total)', 'Standard error of median'], axis=1)
```

```python
combined_df = pd.concat([male_numbers_df, female_numbers_df], 1)
combined_df
```

```python

import matplotlib.pyplot as plt

combined_df.plot(kind='bar', x="Occupation", y=["Total Number of Males", "Total Number of Females"], 
                 title ='Occupational diversity: Males Vs Females', alpha=0.8, align="center", figsize=(18,10),
                 color = ("red", "blue"), fontsize=14)

plt.xlabel("Occupation Class")
plt.ylabel("Males and Females occupation numbers")

plt.show()
```

```python
rename_male_earnings = final_men_df.rename(columns={"Median Earnings(total)":"Male earnings"})
rename_male_earnings.drop(['Total Number of Males', 'Standard error of median'], axis=1)

male_earnings_df = rename_male_earnings["Male earnings"].str.replace(",","").astype(int)
male_earnings_df


rename_female_earnings = final_female_df.rename(columns={"Median Earnings(total)":"Female earnings"})
rename_female_earnings.drop(['Occupation','Total Number of Females', 'Standard error of median'], axis=1)


female_earnings_df = rename_female_earnings["Female earnings"].str.replace(",","").astype(int)
female_earnings_df

compared_earnings_df = pd.concat([combined_df, male_earnings_df, female_earnings_df], 1)
compared_earnings_df

#final_male_earnings_df = pd.DataFrame(male_earnings_df).rename(columns={"Male earnings":"Male_earnings"})

#compared_earnings_df["Female earnings"].str.replace(",","").astype(int)
#final_female_earnings_df = pd.DataFrame(compared_earnings_df).rename(columns={"Female earnings":"Female_earnings"})


#compared_earnings_df = pd.concat([final_male_earnings_df, final_female_earnings_df], 1)

compared_earnings_df.drop(['Total Number of Males', 'Total Number of Females'], axis=1)
```

```python

import matplotlib.pyplot as plt

color = ["red", "green"]

compared_earnings_df.plot(kind='bar', x="Occupation", y=["Male earnings", "Female earnings"], 
                 title ='Earnings: Males Vs Females',
                 alpha=0.6, align="center", figsize=(18,10),color = ("indigo", "orange"),
                         fontsize=14)

plt.xlabel("Occupation Class")
plt.ylabel("Males and Females earnings")

plt.show()
```
